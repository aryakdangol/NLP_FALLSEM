```python
def preprocess(items):
    clean_text=[]
    for item in items:
        #item = remove_emoji(item) #remove emojis
        item = item.lower()
        item = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', '', item) #remove urls
        item = re.sub('@[^\s]+', 'AT_USER', item) #remove usernames
        item = re.sub(r'#([^\s]+)', r'\1', item)  #remove hashtags
        item = remove_stopwords(item)
        item = remove_punct(item)
        clean_text.append(item)
    return clean_text
```

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
def remove_stopwords(string):
    tokens = word_tokenize(string)
    nostopwords = []
    stop_words = set(stopwords.words('english') + ['AT_USER'])
    for word in tokens:
        if word not in stop_words:
            nostopwords.append(word)
    return ' '.join(nostopwords)
```

```python
def remove_punct(string):
    tokens = word_tokenize(string)
    removepunct = []
    punct = set(punctuation)
    for word in tokens:
        if word not in punct:
            removepunct.append(word)
    return  ' '.join(removepunct)
```

```python
#Building Vocabulary
cleantrain = dftrain['Tweet']
labels = dftrain['Label']
cleantrain_labels = zip(cleantrain,labels)
cleantrainData = tuple(cleantrain_labels)
print(cleantrainData[0:5])
```

```python
def buildVocabulary(cleantrainData):
    all_words = []
    for (words,label) in cleantrainData:
        tokens = word_tokenize(words)
        all_words.extend(tokens)
        
    vocabfreq = nltk.FreqDist(all_words)
    word_features = vocabfreq.keys()
    
    return word_features
    
    
```

```python
def extract_features(tweet):
    words = set(tweet)
    features = {}
    for word in word_features:
        features['contains(%s)' % word] = (word in words)
    return features
```

```python
word_features = buildVocabulary(cleantrainData)
trainingfeatures = nltk.classify.apply_features(extract_features, cleantrainData)
```
